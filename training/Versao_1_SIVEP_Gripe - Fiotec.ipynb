```python
path = r'SIVEP-Gripe' # Lembrar de alterar o caminho no notebook de vocês

files = glob.glob(path + "/*.csv")

li = []

# Realiza a leitura de todos os conjuntos de dados disponiveis no diretório
# Aqui assumimos que todos possuem a mesma estrutura de colunas
# Se houver discordância o código não irá funcionar.
# Criamos uma lista "li" que irá receber os conjuntos
for filename in files:
  print(filename)
  df = pd.read_csv(filename,
                   sep=';', 
                   encoding='utf-8', 
                   error_bad_lines=False)
  li.append(df)


# Essa lista é concatenada e vira um só banco de dados

dados_rs = pd.concat(li, axis=0, ignore_index=True)

# Converte todas as colunas pra minúsculo. Coisa de preguiçoso pra não dar o caps lock toda vez que for fazer algo com as colunas rsrs

dados_rs.columns= dados_rs.columns.str.lower()
```

```python
# Variáveis de configuração do funcionamento do código

COLUNAS = ['estadonotificacao', 'municipionotificacao',
           'sexo', 'racacor', 'estado', 'municipio',
           'datanotificacao', 'sintomas', 'datainiciosintomas',
           'condicoes', 'evolucaocaso', 'classificacaofinal', 'idade']

COLUNAS_DATAS = ['datanotificacao','datainiciosintomas']

BINS_IDADE = [-1,19,39,49,59,69,79,140]
```

```python
# Funções úteis

def qnt_condicoes(condicao):
  # print(condicao)
  if type(condicao) is str:
    return len(condicao.split(','))
  else:
    return 0
```

```python
dados_rs = dados_rs[COLUNAS]
```

```python
# Mantém apenas casos confirmados
dados_rs = dados_rs[dados_rs['classificacaofinal']!='Descartado']
dados_rs = dados_rs[dados_rs['classificacaofinal']!='Síndrome Gripal Não Especificada']

# Elimina os dados nulos, sem diagnostico final
dados_rs = dados_rs.dropna(subset='classificacaofinal')

# Remove casos sem data de notificacao válida
dados_rs = dados_rs[dados_rs['datanotificacao'] != 'None']
dados_rs = dados_rs[dados_rs['datainiciosintomas'] != 'None']

# Mantém apenas os registros de pessoas notificadas e domiciliadas no RS
# Filtro por garantia
dados_rs = dados_rs[dados_rs['estadonotificacao']=='Rio Grande do Sul']
dados_rs = dados_rs[dados_rs['estado']=='Rio Grande do Sul']

# Mantém apenas individuos do sexo Feminino e Masculino
dados_rs = dados_rs[dados_rs['sexo'] != 'Indefinido']
dados_rs = dados_rs.dropna(subset=['sexo'])
```

```python
# Preenche os campos sem Raça/Cor como ignorados
dados_rs['racacor'] = dados_rs['racacor'].fillna('Ignorado')
```

```python
# Ajusta as datas do conjunto de dados
# Processa as colunas com datas, convertendo o tipo de dados Objeto para Datetime

dados_rs[COLUNAS_DATAS] = dados_rs[COLUNAS_DATAS].apply(pd.to_datetime, format='%Y/%m/%d')
```

```python
# Verifica a quantidade de dados faltantes para as datas

dados_rs[COLUNAS_DATAS].isnull().sum()
```

```python
# Cria a semana de notificacao
dados_rs['sem_not'] = dados_rs['datainiciosintomas'].dt.isocalendar().week
```

```python
# Cria categorias etárias
dados_rs['faixa_etaria'] = pd.cut(dados_rs['idade'], bins=BINS_IDADE)
```

```python
# Cria uma lista para auxiliar nos plots
labels_faixa = []

for c in dados_rs['faixa_etaria'].cat.categories:
  lower_val, upper_val = c.left, c.right
  labels_faixa.append('%s para %s'%(lower_val+1, upper_val))

print(labels_faixa)
labels_faixa[-1] = str(BINS_IDADE[-2]+1)+' <'
print(labels_faixa)
```

```python
# Cria contagem de dias ate notificacao
dados_rs['dias_atenot'] = (pd.to_datetime(dados_rs['datanotificacao'], format="%Y-%m-%d") - pd.to_datetime(dados_rs['datainiciosintomas'],format="%Y-%m-%d")).dt.days
print("Dias até notificação: max {} e min {}".format(dados_rs['dias_atenot'].max(),dados_rs['dias_atenot'].min()))
```

```python
# Remove casos com mais de 365 dias e menos que 0 dias
dados_rs = dados_rs[(dados_rs['dias_atenot'] > 0)& (dados_rs['dias_atenot'] < 365)]
```

```python
# Visão geral de dados faltantes para o conjunto de dados 
# após o tratamento de dados
dados_rs.isnull().sum()
```

```python
# # Fonte https://infoms.saude.gov.br/extensions/covid-19_html/covid-19_html.html
# print("O percentual de casos considerados para o treinamento em relação a\nquantidade de casos confirmados no Estado do RS para o ano de 2021 é:")
# print("{:.2f}%".format((dados_rs.shape[0]/1507117)*100))
```

```python
dados_rs['qnt_sintomas'] = dados_rs['sintomas'].apply(qnt_condicoes)
dados_rs['qnt_comorbidades'] = dados_rs['condicoes'].apply(qnt_condicoes)
```

```python
# Pega os óbitos

obitos_rs = dados_rs[dados_rs['evolucaocaso'] == 'Óbito']
```

### Verificando as colunas do Dataset

```python
all_muni = geobr.read_municipality(code_muni="RS", year=2020)
# Remove os últimos 3 digitos para dar match com o cod da SRAG
all_muni['code_muni'] = all_muni['code_muni'].astype(str).str[:-3].astype(np.int64)
```

```python
pop_rs = pd.read_csv('pop_rs_fixed.csv', encoding='utf-8', sep=',', usecols=['Municipio', 'Codigo', 'Populacao'])

# Remove os últimos dois digitos para dar match com o cod da SRAG
pop_rs['Codigo'] = pop_rs['Codigo'].astype(str).str[:-1].astype(np.int64)
```

```python
# Converte o nome de todos os municipios para minusculo
dados_rs['municipio'] = dados_rs['municipio'].str.lower()
pop_rs['Municipio'] = pop_rs['Municipio'].str.lower()
all_muni['name_muni'] = all_muni['name_muni'].str.lower()
```

```python
pop_rs
```

```python
# Conta a quantidade de casos por COD do munícipio de residencia
acum_casos = dados_rs['municipio'].value_counts().rename_axis('municipio').reset_index(name='casos')

# Faz um left join dos casos com os dados da populacao
all_muni = all_muni.merge(acum_casos, how="left", left_on="name_muni", right_on="municipio")
all_muni = all_muni.merge(pop_rs, how="left", left_on="name_muni", right_on="Municipio")
```

```python
# Calcula a quantidade de casos hospitalizados por 1000 habitantes por municipio de residencia
all_muni['casos1000'] = (all_muni['casos'] / all_muni['Populacao']) * 1000
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

all_muni.plot(
    column="casos1000",
    cmap="Blues_r",
    legend=True,

    
    legend_kwds={
        "label": "Casos por 1000 habitantes",
        "orientation": "horizontal",
        "shrink": 0.6,
    },
    ax=ax,
)

ax.set_title("Casos de COVID-19 hospitalizados por 1000 habitantes, Cidades RS, 2021", fontsize=10)
ax.axis("off")
```

Analisa a piramide etária

```python
# Realiza a contagem de casos por sexo e faixa etaria
piramide_etaria = dados_rs[['sexo','faixa_etaria']].value_counts().rename_axis(['sexo','faixa_etaria']).reset_index(name='casos')

# Inverte os valores para o sexo Masculino
piramide_etaria.loc[piramide_etaria['sexo'] == 'Masculino', 'casos'] = 0 - piramide_etaria.loc[piramide_etaria['sexo'] == 'Masculino', 'casos']

# Cria uma coluna para cada sexo
piramide_etaria = piramide_etaria.pivot_table('casos', ['faixa_etaria'], 'sexo')
piramide_etaria = piramide_etaria.reset_index()

# Ajustas as faixas etárias
piramide_etaria['faixa_etaria'] = labels_faixa
```

```python
plt.figure(figsize=(8,6))

bar_plot = sns.barplot(x='Masculino', y='faixa_etaria', data=piramide_etaria, order=labels_faixa[::-1], color='b', label='Masculino')
bar_plot = sns.barplot(x='Feminino', y='faixa_etaria', data=piramide_etaria, order=labels_faixa[::-1], color='r', label='Feminino')
bar_plot.set(xlabel='Casos', ylabel="Faixa Etária", title='Piramide Populacional de Casos de COVID-19 hospitalizados no RS - 2021')

# plt.xticks(ticks=[-12500,-10000,-7500,-5000,-2500,0,2500,5000,7500,10000,12500],
          #  labels=['12500','10000','7500','5000','2500','0','2500','5000','7500','10000','12500'])

plt.legend(loc='lower right')
plt.show()
```

```python
# Casos ao longo de cada semana epidemiológica
hist_casos = dados_rs['sem_not'].value_counts().rename_axis(['semana']).reset_index(name='casos')
hist_obitos = dados_rs[dados_rs['evolucaocaso'] == 'Óbito']['sem_not'].value_counts().rename_axis(['semana']).reset_index(name='casos')


plt.figure(figsize=(15,6))
bar_plot = sns.barplot(data=hist_casos, x='semana', y='casos', color='b', label='Casos')
bar_plot = sns.barplot(data=hist_obitos, x='semana', y='casos', color='r', label='Óbitos')

bar_plot.set(xlabel='Semana Epidemiológica de 2021', ylabel='Casos', title='Casos de COVID-19 hospitalizados no RS por semana epidemiológica - 2021')

plt.legend(loc='upper right')
plt.show()
```
